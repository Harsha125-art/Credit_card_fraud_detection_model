## Credit Card Fraud Detection Model 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score
print("setup completed")
```

## 1. Introduction  
- Detect fraudulent transactions in a highly imbalanced dataset is the main objective
### Dataset contains:
- 284,807 transactions, only 492 frauds  


```python
df = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv")
df.head()
```

#### All the columns named v1,v2.. to not to protect privacy as these columns include credit card details 

```python
df.info()
```

```python
df.describe()
```

```python
df['Class'].value_counts()
```

## 2. Data Preprocessing  
- Removed `Class` from training features to avoid data leakage  


```python
X = df.drop('Class',axis=1)
y = df['Class']

X_train ,X_valid,y_train,y_valid = train_test_split(X,y,test_size=0.2,stratify=y,random_state=1)
```

## 3. Model Building  
- Used Random Forest  
- Evaluated with Confusion Matrix, Precision, Recall, ROC AUC  


```python
model_1 = RandomForestClassifier(n_estimators=100,random_state=1)
model_1.fit(X_train,y_train)

y_predict_1=model_1.predict(X_valid)

print(confusion_matrix(y_valid,y_predict_1))
print(classification_report(y_valid,y_predict_1))
print("Roc Auc :",roc_auc_score(y_valid,model_1.predict_proba(X_valid)[:,1]))

```

```python
from sklearn.metrics import roc_curve ,auc

y_pred = model_1.predict_proba(X_valid)[:,1]

fpr,tpr,thresholds = roc_curve(y_valid,y_pred)
roc_auc = auc(fpr,tpr)

plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,color='orange')
plt.plot([0,1],[0,1],color='navy',linestyle='--')
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.grid(True)
plt.show()
print(roc_auc)
```

## 4. Threshold Tuning  
- Custom threshold = 0.3  
- Increased Recall to catch more frauds  


```python
custom_threshold = 0.3
y_predict_2 = (y_pred>custom_threshold).astype(int)


print(confusion_matrix(y_valid,y_predict_2))
print(classification_report(y_valid,y_predict_2))
```

## 5. Improving Model Performance  
- Using XGBoost to increase performance
- ROC AUC: 0.9826  
- Precision: 0.92, Recall: 0.86  


```python
from xgboost import XGBClassifier

model_2 = XGBClassifier(n_estimators=100,learning_rate=0.1,random_state=1)
model_2.fit(X_train,y_train)
model_2_predict=model_2.predict(X_valid)

print(confusion_matrix(y_valid,model_2_predict))
print(classification_report(y_valid,model_2_predict))
print("Roc Auc :",roc_auc_score(y_valid,model_2.predict_proba(X_valid)[:,1]))
```

```python
y_pred_new = model_2.predict_proba(X_valid)[:,1]

fpr,tpr,thresholds = roc_curve(y_valid,y_pred_new)
roc_auc = auc(fpr,tpr)

plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,color='orange')
plt.plot([0,1],[0,1],color='navy',linestyle='--')
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.grid(True)
plt.show()
print(roc_auc)
```

## 5. Final Model Performance  
- ROC AUC: 0.9826  
- Precision: 0.91, Recall: 0.89  


```python
custom_threshold = 0.3
y_predict_3 = (y_pred_new>custom_threshold).astype(int)



print(confusion_matrix(y_valid,y_predict_3))
print(classification_report(y_valid,y_predict_3))

```
